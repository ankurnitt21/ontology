# Checking similarity between words in a pair for the relation "HAS_SYNONYM"

```python
syno=[]
for word in pair:
    l1=[]
    l2=[]
    max1=0.0
    for synset in wn.synsets(str(word[0])):
        for lemma in synset.lemmas():
            if lemma.name() not in l1:
                l1.append(str(lemma.name()))
    for synset in wn.synsets(str(word[1])):
        for lemma in synset.lemmas():
            if lemma.name() not in l2:
                l2.append(str(lemma.name()))
    for x in l1:
        syn1=wordnet.synsets(x)[0]
        for y in l2:
            syn2=wordnet.synsets(y)[0]
            if (syn1.wup_similarity(syn2))!=None :
                 max1=max(max1,syn1.wup_similarity(syn2))
    if max1==1.0:
        syno.append(word)
print(syno)
```
